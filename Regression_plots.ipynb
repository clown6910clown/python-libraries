# Regression Plots

Regression is for numerical data, Classification is for categorial data

```python
import seaborn as sns
%matplotlib inline
```

```python
tips= sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)  #lmplot is linear model plot.
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='time')  #hue gives color to all unique parameters
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],palette='coolwarm')
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100},palette='plasma')
#scatter_kws is used to increase or decrease the size of the point
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':780},palette='plasma')
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':5},line_kws={'lw':10},palette='plasma')
#line_kws is used to change the line width.
```

```python

sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='smoker',aspect=0.6,palette='coolwarm')
```

```python
res= sns.kdeplot(data=tips,x=tips['total_bill'],y=tips['tip'],fill="green")
```
