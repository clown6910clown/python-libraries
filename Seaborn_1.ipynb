#Seaborn

```python
import seaborn as sns #import seaborn

sns.set_theme() #apply the default theme

#load an example dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

plotting

```python
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
import numpy as np
```

countplot

```python
sns.countplot(x="day", data=tips)
```

Boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette="rainbow")
```

```python
#can do entire dataframe with orient='h', it will give horizontally

sns.boxplot(data=tips,palette="rainbow",orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips,palette='coolwarm')
```

violin plot

```python
#it shows the distribution of quantitative data

sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1')
```

informative distribution summary

```python
sns.displot(data=tips,x='total_bill',col="time",kde=True) #kde is kernerl distribution estimator, displot gives distribution
```

```python
sns.distplot(tips['total_bill']) #distribution of total bills from the tips
```

```python
sns.distplot(tips['total_bill'],kde=True,bins=5)
```

```python
sns.distplot(tips['total_bill'],kde=True,bins=30)
#to remove the kde layer and just have histogram use
```

jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',color='red')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

Pairplots

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

rugplot

```python
sns.rugplot(tips['day'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
