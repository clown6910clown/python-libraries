Mathematical Operations

```python
result = stats.BirthRate * stats.InternetUsers
print(result.head())
```

```python
stats['MyCal']=stats.BirthRate * stats.InternetUsers
print(stats.head()) #adding the mathematical operation as column
```

```python
#Removing the column
stats=stats.drop('MyCal',axis=1) #creates a new object by defalt unless you assign it
stats.head()
```

```python
Filter=stats['InternetUsers']<2 #filtering based on internet users
Filter
stats[Filter]
```

```python
#filtering based on birthrate
Filter2=stats['BirthRate']>40
stats[Filter2]
```

```python
#Applying 2 Filters
stats[Filter & Filter2]
```

```python
#applying 2 flters using square bracket approach
stats[(stats.BirthRate>40)&(stats.InternetUsers<2)] #need parenthess for orde of operations
```
