Most Common Scaling Techniques are Normalization, Standardization


*   Normalization is also called Min-Max Scaling. it is used to transform features to be on a similar scale. this scale is in rage [0,1] or [-1,1]

*   Standardization or Z score Normalization is transformation of features by substracting from mean and dividing by standard deviation





```python
import pandas as pd
import numpy as np
```

# 1.Min_Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler #importing min-max scaler (normalization scaler)
```

```python
#define data

data= pd.DataFrame(data=[[5,5],[10,2],[2,0],[0,10]],columns=['X','Y'])
data
```

```python
#plot scatter plot

import matplotlib.pyplot as plt
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c="red")
plt.axhline(y=0,c="yellow")
plt.show()
```

Define, fit and transofrm min-max scaler

```python
scaler= MinMaxScaler() # y=(x-min)/(max-min)

min_max_scaled= scaler.fit_transform(data)
```

```python
#print the scaled values
min_max_scaled=pd.DataFrame(min_max_scaled,columns=['X','Y'])
min_max_scaled
```

```python
import matplotlib.pyplot as plt
plt.scatter(min_max_scaled.X,min_max_scaled.Y)
plt.axvline(x=0,c="red")
plt.axhline(y=0,c="yellow")
plt.show()
```

### Sub plots

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c="red")
plt.axhline(y=0,c="yellow")

plt.subplot(1,2,2)
plt.scatter(min_max_scaled.X,min_max_scaled.Y)
plt.show()
```

# Standardization

Standard scaler

```python
#import standard scaler
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.DataFrame(data=[[5,5],[10,2],[2,0],[0,10]],columns=['X','Y'])
data
```

```python
scaler1 = StandardScaler()  #X new = (X- X mean) / (X std)
std_scaled= scaler1.fit_transform(data)
std_scaled
```

```python
#print the scaled values
std_scaled=pd.DataFrame(std_scaled,columns=['X','Y'])
std_scaled
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.scatter(std_scaled.X,std_scaled.Y)
plt.axvline(x=0,c="red")
plt.axhline(y=0,c="yellow")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))

plt.subplot(1,2,2)
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c="red")
plt.axhline(y=0,c="yellow")
plt.show()

plt.subplot(1,2,2)
plt.scatter(std_scaled.X,std_scaled.Y)
plt.axvline(x=0,c="red")
plt.axhline(y=0,c="yellow")
plt.show()
```

###plot both min_max and standard scaler

```python
import matplotlib.pyplot as plt
plt.scatter(min_max_scaled.X,min_max_scaled.Y,marker='^')
plt.scatter(std_scaled.X,std_scaled.Y,marker='o')
plt.axvline(x=0,c="red")
plt.axhline(y=0,c="yellow")
plt.show()
```
