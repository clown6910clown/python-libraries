```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y,'r') # 'r' is the color red
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1) #plt.subplot(nrows,ncols,plot_number)
plt.plot(x,y,'r--') #More on Colour Options

plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
#Creating Figure (Empty Canvas)
fig=plt.figure()

#adding Set of Axes to Figure
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height (range 0 to 1)

#Plotting Data
axes.plot(x,y,'b') #'b' for color blue
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
#Creating blank Canvas
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #Main Axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #Inset Axes

#Larger Figures Axes 1
axes1.plot(x,y,'b')
axes.set_xlabel('X Label Axes 1')
axes.set_ylabel('Y Label Axes 2')
axes.set_title('Axes 1 Title')

#Insert figure Axes 2
axes2.plot(x,y,'r')
axes.set_xlabel('X Label Axes 2')
axes.set_ylabel('Y Label Axes 2')
axes.set_title('Axes 2 Title')
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,'r') #'b' for color blue
axes.set_xlabel('X ')
axes.set_ylabel('Y ')
axes.set_title(' Title')
```

## Figure Size, Aspect Ratio and dpi

```python
fig, axes= plt.subplots(nrows=1,ncols=2)

for ax in axes:
  ax.plot(x,y,'b')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('title')
fig
plt.tight_layout()
```

```python
fig, axes=plt.subplots(figsize=(8,4))


axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig= plt.figure(figsize=(8,4),dpi=1000)
fig, axes=plt.subplots(figsize=[12,3])
axes.plot(x,y)
fig.axes[0].set_xlabel('x')
fig.axes[0].set_ylabel('y')
fig.axes[0].set_title('title')
```

## Labels, legends and titles

```python
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend()
```

```python
#we have many options

ax.legend(loc=1) #upper right corner
ax.legend(loc=2) #upper left corner
ax.legend(loc=3) #lower left corner
ax.legend(loc=4) #lower right corner

#and we also have other options

#most common choice is
ax.legend(loc=0)  #matplot lib will decide the optimal location
fig
```

## setting colors

```python
fig, ax = plt.subplots()

ax.plot(x,x+1,color="blue",alpha=0.5) #half transparency
ax.plot(x,x+2,color="#8B008B") #RGB Hex Code
ax.plot(x,x+3,color="#FF8C00") #RGB Hex Code
```

## Line and Width

```python
fig, ax= plt.subplots(figsize=(12,6))

ax.plot(x,x+1,color="red",linewidth=0.25)
ax.plot(x,x+2,color="red",linewidth=0.50)
ax.plot(x,x+3,color="red",linewidth=1.00)
ax.plot(x,x+4,color="red",linewidth=2.00)

#possible linestyle type options '-','-.',':'

ax.plot(x,x+5,color="green",lw=3,linestyle='-')
ax.plot(x,x+6,color="green",lw=3,linestyle='-.')
ax.plot(x,x+7,color="green",lw=3,linestyle=':')
#custom dash
line, = ax.plot(x,x+8,color="black",lw=1.50)
line.set_dashes([5,10,15,10]) #format line length,space length

#possible marker symbols:marker='+','o','*','s',',','.','1','2','3','4',
ax.plot(x,x+9,color="blue",lw=3,ls='-',marker='+')
ax.plot(x,x+10,color="blue",lw=3,ls='--',marker='o')
ax.plot(x,x+11,color="blue",lw=3,ls='-',marker='s')
ax.plot(x,x+12,color="blue",lw=3,ls='--',marker='1')

#marker size and color

ax.plot(x,x+13,color="purple",lw=1,ls='-',marker='o',markersize=2)
ax.plot(x,x+14,color="purple",lw=1,ls='-',marker='o',markersize=4)
ax.plot(x,x+15,color="purple",lw=1,ls='-',marker='o',markersize=8,markerfacecolor='red')
ax.plot(x,x+16,color="purple",lw=1,ls='-',marker='s',markersize=8,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```

Plot Ranges

```python
fig, axes=plt.subplots(1,3,figsize=(12,4)) #this means we will have 3 pics in 1 column with aspect ratio 12,4

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('default axes ranges')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('custom axes ranges')
```
