The grid() function of axes objects sets visibility of grid inside the figure
*   It helps to understand the data better
*   It acts like a scale



```python
import matplotlib.pyplot as plt
import numpy as np
fig, axes= plt.subplots(1,3,figsize=(12,4))
x=np.linspace(1,11)

axes[0].plot(x,x**3,'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')

axes[1].plot(x,np.exp(x),'r',lw=2)
axes[1].grid(color='b',ls='--',lw=0.25)
axes[1].set_title('custom grid')

axes[2].plot(x,x)
axes[2].set_title('no grid')
fig.tight_layout() #this is used to avoid overlapping of images
plt.show()
```

Spine is the way we define the outer edge of the canvas

```python
import matplotlib.pyplot as plt
fig= plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.grid("True")
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color('black')
ax.spines['top'].set_color('green')
ax.plot([5,6,7,8,1])
plt.show()
```

```python
#using iris dataset
import matplotlib.pyplot as plt
import seaborn as sns
#to generate output in same window
%matplotlib inline
iris=sns.load_dataset('iris')
iris.head()
```

```python
#Pair  plot :- every feauture will be compared with every other feature

sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris,hue='species',palette='rainbow')
```

```python
sns.PairGrid(data=iris) #it just marks the grids
```

```python
g= sns.PairGrid(iris) #then you map to the grid, Pair Grid
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot) #kernel density estimator plot
```

```python
#use tips dataset
tips=sns.load_dataset('tips')
tips.head()
```

Facet Grid is a general way to creating plot grids based on a function

```python
g= sns.FacetGrid(tips,col="time",row="smoker")
```

```python
g=sns.FacetGrid(tips,col="time",row="smoker")
g=g.map(plt.hist,'total_bill')
#Who has paid the maximum bill during dinner? smoker or non-smoker?
```

```python
g=sns.FacetGrid(tips,col="time",row="smoker",hue='sex',palette="coolwarm")
g=g.map(plt.scatter,'total_bill',"tip").add_legend()
```

```python
g=sns.JointGrid(x='total_bill',y='tip',data=tips)
```

```python
g=sns.JointGrid(x='total_bill',y='tip',data=tips)
g=g.plot(sns.regplot,sns.distplot)
```

Grids in Matplotlib

```python
import matplotlib.pyplot as plt #implementation of matplot function
import numpy as np
#dummy data
x1=np.linspace(0.0,5.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)

print(x1)
print(y1)
#creates two subplots
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,5))

#plot without a grid

ax1.plot(x1,y1)
ax1.set_title('plot without grid')

#plot with a grid
ax2.plot(x1,y1)
ax2.set_title('plot with a grid')
#draw gridlines
ax2.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#dummy data
x=np.linspace(0,2*np.pi,400) #400 means we need 400 values
y=np.sin(x**2)
#set graph color
plt.plot(x,y,'green')
#to set title
plt.title("Plot with linewidth and linestyle")
#draw gridlines grey using given linewidth and linestyle

plt.grid(True,color='grey',linewidth='0.4',linestyle=":")
plt.show()
```
