Correlation Matrix

```python
tips=sn.load_dataset("tips") #loading the tips dataset using seaborn
tips.head()
```

Correlation matrix can be used to see the dependency between two variables

*   Condition: Both have to be numerical matrix
*   it states that if one variable changes, the other one also changes



```python
tips.corr(numeric_only=True) #Matrix form for correlation data , we also need to put numeric_only as true to only take numerical values
```

```python
sn.heatmap(tips.corr(numeric_only=True))
```

```python
sn.heatmap(tips.corr(numeric_only=True),cmap='coolwarm',annot=True)
```
